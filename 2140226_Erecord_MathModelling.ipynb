# SOLUTION OF INITIAL VALUE PROBLEM (IVP):

# a) $ dy/dx=2k, y(0)=1$

```python
from numpy import *
from matplotlib.pyplot import *
from scipy.integrate import odeint
plt.figure(figsize=(6,6))
plt.style.use("seaborn-whitegrid")
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.axhline(0,color="black")
plt.axvline(0,color="black")
plt.xlabel("x-label",fontsize=14)
plt.ylabel("y-label",fontsize=14)
def model(y,t,k):
    dydx=2*k
    return dydx
y0=1
t=linspace(0,20,50)
#solve ODE
k=1
yk1=odeint(model,y0,t,args=(k,))
k=2
yk2=odeint(model,y0,t,args=(k,))
k=5
yk3=odeint(model,y0,t,args=(k,))
plot(t,yk1,'r+',label='k=1')
plot(t,yk2,'b-',label='k=2')
plot(t,yk3,'g*',label='k=3')
plt.legend(fontsize=10,loc='upper left')
plt.show()
```

# b) $ dy/dx=5y $

```python
from numpy import *
from matplotlib.pyplot import *
from scipy.integrate import odeint

plt.figure(figsize=(6, 6))
plt.style.use("seaborn-whitegrid")
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.axhline(0, color="black")
plt.axvline(0, color="black")
plt.xlabel("x-label", fontsize=14)
plt.ylabel("y-label", fontsize=14)

def model(y, t):
    dydx = 5 * y
    return dydx

t = linspace(0, 5, 100)  # Time domain from 0 to 5

y = odeint(model, 1, t)  # Solving the ODE with y0 = 1

plot(t, y, 'b-', label='dy/dx = 5y')
plt.legend(fontsize=10, loc='upper left')
plt.show()
```

## 1. Cells in a dish have started to decay exponentially by 3% every 20 minutes. When you left there were 1000 cells in the dish. Now there are 92 left.Approxiamately how long did you leave the cell for?

```python
from sympy import *
t,k,C1=symbols("t,k,C1")
p=Function("p")(t)
```

```python
# The differential equation is given by:
DE=Eq(diff(p,t),-(3/2000)*p)
DE
```

```python
# The general solution of the above DE is given by:
GS=dsolve(DE,p)
GS
```

```python
# At time t=0,p(0)=1000
s1=GS.subs({t:0,p:1000})
s1
```

```python
# Substitute the value od c1 in the GS
s2=GS.subs(C1,1000)
s2
```

```python
s3=s2.subs(p,92)
s3
```

```python
w=solve(s3,t)
d=[n for n in w if n.is_real]
d
```

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0,2000,1000)
p = 1000*np.exp(-(3/2000)*t) 

plt.plot(t,p)
plt.xlabel('Time (min)')
plt.ylabel('Cells remaining')
plt.title('Exponential Decay of Cells')
plt.show()
```

## 1. The half-life of an element is 186 days. If 68g of the element was present in the lab, what fraction of the sample remains after a year?

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
t,k=sp.symbols("t,k")
p=Function("p")(t)
e=sp.Eq(p.diff(t),-k*p)
display("The differential equation is:",e)
de=sp.dsolve(e)
de1=de.subs({t:0,p:68})
display(de1)
C1=sp.Symbol("C1")
de2=de.subs(C1,68)
display(de2)
de3=de2.subs({t:186,p:34})
display(de3)
w=sp.solve(de3)
display(w)
w1=[n for n in w if n.is_real]
display(w1)
```

## PHASE PORTRAIT OF A SYSTEM OF ODEs:
<b>A phase portrait is a geometric representationof the orbits of a dynamnical system in the phase plane. Each set of initial conditions is represented by a different point or curve</b>
1) Phase port are an invaluable tool in studying dynamical system.<br>
2) They consist of a plot of typical trajectoriesin the phase space.This reveals informationsuch as whether an attractor, a repellor or limit cycle is present for the chosen parameter value.<br>
3) The concept of topological equivalenceis important in classifyingthe behaviour of systems by specifying when two different phase portraits represent the same qualitative dyanamic behaviour.<br>
4) An attractor is a stable point which is also called a "sink". The repellor is considered as an unstable point, which is also known as a "source".<br>
5) A phase portrait graph of a dynamical system depicts the system's trajectories(with arrows) and stable steady states (with dots) and unstable steady states  (with circles) in a phase space.


## An undamped pendullum with no driving force id described by:
## $y'' + sin(y)=0$
## We reduce this system in the first order ODEsby letting $ y_1=y$ and $ y_2=y_1'$.This leads to:
## $ y_1'=y_2$
## $y_2'=-sin(y)$
<b> The phase portrait is a plot of a vector field which qualitatively shows how the solution to these equations will go from a given starting point. Here is our definition of the different equations:<br>
    To generate these phase portraits, we need to compute the derivatives$ y_1'$ and $ y_2'$at $ t=0$ on a grid over the range of values for $ y_1$ and $ y_2 $ we are interested in. We will plot the derivatives as a vector of each $(y_1,y_2)$which will show the initial direction from each point. We will examine the solution over the range $ -2<y_1<8$ and  $ -2<y_2<2$ for $ y_2$ and create a grid of 20*20 points.

```python
import numpy as np
import matplotlib.pyplot as plt
def f(y,t):
    y1,y2=y
    return [y2,-np.sin(y1)]

y1=np.linspace(-2,8,20)
y2=np.linspace(-2,2,20)

Y1,Y2=np.meshgrid(y1,y2)

t=0

u,v=np.zeros(Y1.shape),np.zeros(Y2.shape)
NI,NJ=Y1.shape

for i in range(NI):
    for j in range(NJ):
        x=Y1[i,j]
        y=Y2[i,j]
        yprime=f([x,y],t)
        u[i,j]=yprime[0]
        v[i,j]=yprime[1]
Q=plt.quiver(Y1,Y2,u,v,color='r')

plt.xlabel('$y_1$')
plt.ylabel('$y_2$')
plt.xlim([-2,8])
plt.ylim([-4,4])
```
